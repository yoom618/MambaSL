```python
# generate UCR yaml config files for all datasets in a given directory

import os

data_dir = '/data/yoom618/TSLib/dataset_UCR'

configs = {}
for dataset in os.listdir(data_dir):
    dataset_path = os.path.join(data_dir, dataset)
    if os.path.isdir(dataset_path):
        print(f'Processing dataset: {dataset}')

        with open(f'/data/yoom618/TSLib/dataset_UCR/{dataset}/{dataset}_TRAIN.ts', 'r') as f:
            content = filter(lambda x: (x[0] == '@') & (x[:5] != '@data'), f.readlines())
            content = list(map(lambda x: x.strip().split(' '), content))
            content = dict(map(lambda x: (x[0][1:], x[1:]), content))

        configs[dataset] = {
            'task_name': 'classification',
            'dataset': dataset,
            'data': 'UEA',
            'root_path': os.path.join(data_dir, dataset),
            # Placeholder values; these would need to be set based on actual dataset properties
            'seq_len': int(content['seriesLength'][0]) if content['equalLength'][0] == 'true' else 0,
            'enc_in': 1,
            'num_class': len(content['classLabel']) - 1,
            'label_len': 0,
            'pred_len': 0,
            'c_out': 0,
        }
        print(f'Generated config for {dataset}')
        if content['missing'][0] == 'true':
            print(f'\tWarning: Dataset {dataset} has missing values.')
        if content['equalLength'][0] == 'false':
            print(f'\tWarning: Dataset {dataset} has variable length sequences.')
    
# yaml_file = os.path.join('./data_classification_ucr_tmp.yaml')
# with open(yaml_file, 'w') as f:
#     yaml.dump(configs, f)
# print(f'Generated config for {dataset}: {yaml_file}')
```
