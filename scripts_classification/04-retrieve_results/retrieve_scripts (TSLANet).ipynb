```python
import os
import pandas as pd
```

```python
UEA_MTSC30 = ['EthanolConcentration',
              'FaceDetection',
              'Handwriting',
              'Heartbeat',
              'JapaneseVowels',
              'PEMS-SF',
              'SelfRegulationSCP1',
              'SelfRegulationSCP2',
              'SpokenArabicDigits',
              'UWaveGestureLibrary',
              'ArticularyWordRecognition',
              'AtrialFibrillation',
              'BasicMotions',
              'CharacterTrajectories',
              'Cricket',
              'DuckDuckGeese',
              'EigenWorms',
              'Epilepsy',
              'ERing',
              'FingerMovements',
              'HandMovementDirection',
              'InsectWingbeat',
              'Libras',
              'LSST',
              'MotorImagery',
              'NATOPS',
              'PenDigits',
              'PhonemeSpectra',
              'RacketSports',
              'StandWalkJump']
```

## make scripts

```python
os.makedirs('TSLANet_scripts', exist_ok=True)

# for data_name in ['EthanolConcentration']:
for data_name in UEA_MTSC30:
    df = pd.read_csv(f'../06-visualize_results/TSLANet/{data_name}.csv')
    log_paths = df[df['acc'] == df['acc'].max()].sort_values('model_size (KB)')['log_path'].values

    scripts = f'''model_name="TSLANet"
dataset_name="{data_name}"
tslib_dir="/data/yoom618/TSLib"
gpu_id=0

data_dir="${{tslib_dir}}/dataset"
checkpoint_dir="${{tslib_dir}}/checkpoints_best/${{model_name}}"\n\n'''
    
    if len(log_paths) > 1:
        scripts += '# below all have the same performance\n\n'
    
    for log_path in log_paths:
        _, dim, depth, mask, patch, mlp_ratio, _, asb, _, af, _, icb, _, pretrain, hh, mm, ss = list(filter(lambda x: x != '', log_path.split('_')))

        script = f'''python -u _run_TSLANet/TSLANet_classification_test.py \\
    --gpu ${{gpu_id}} \\
    --data_type uea \\
    --data_path "${{data_dir}}/${{dataset_name}}" \\
    --data_name ${{dataset_name}} \\
    --ckpt_path ${{checkpoint_dir}} \\
    --ckpt_time "{hh}_{mm}_{ss}" \\
    --model_id "CLS_${{dataset_name}}" \\
    --depth {depth.replace('depth', '')} \\
    --emb_dim {dim.replace('dim', '')} \\
    --mlp_ratio {mlp_ratio.replace('mlp', '')} \\
    --masking_ratio {mask.replace('mask', '')} \\
    --ICB True \\
    --ASB True \\
    --adaptive_filter True \\
    --load_from_pretrained True \\
    --patch_size {patch.replace('patch', '')}\n\n'''

        scripts += script

    # print(scripts[:-1])
    with open(f'TSLANet_scripts/{data_name}.sh', 'w') as f:
        f.write(scripts[:-1])
    
    print(f'Finish writing {data_name}.sh')
```

## move ckpt

```python
import shutil

save_path = '/data/yoom618/TSLib/logs (TSLANet)'  # original log path (--save_path in the scripts)
for data_name in UEA_MTSC30:
    df = pd.read_csv(f'../06-visualize_results/TSLANet/{data_name}.csv')
    log_paths = df[df['acc'] == df['acc'].max()].sort_values('model_size (KB)')['log_path'].values

    for log_dir in log_paths:
        log_dir = os.path.join(save_path, log_dir)
        if os.path.exists(log_dir):
            print(f'move {log_dir} to {log_dir.replace("logs (TSLANet)", "checkpoints_final/TSLANet")}')
            os.makedirs(log_dir.replace("logs (TSLANet)", "checkpoints_final/TSLANet"), exist_ok=True)
            shutil.move(log_dir, log_dir.replace("logs (TSLANet)", "checkpoints_final/TSLANet"))
        else:
            print(f'{log_dir} not exists')
```
