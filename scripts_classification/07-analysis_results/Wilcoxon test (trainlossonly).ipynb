```python
import numpy as np
import pandas as pd
pd.set_option('display.float_format', lambda x: '%.13f' % x)
pd.set_option('display.max_columns', None)

from scipy.stats import wilcoxon
```

```python
df = pd.read_excel('trainlossonly-result.xlsx', index_col=0)
df.head()
```

```python
df_norm = ((df.T - df.iloc[:,7:].mean(axis=1)) / df.iloc[:,7:].std(axis=1)).T
wilcoxon_res = list()
for c in ['CIF (HC2)', 'DrCIF (HC2)', 'HC2 (HC2)',
          'MiniRocket (BORF)', 'DrCIF (BORF)', 'MR+Hydra (BORF)', 
          'MR+Hydra', 'HC2', 'Hydra']:

    valid_df = df_norm[['MambaSL (trainlossonly)', c]].dropna()
    res = wilcoxon((valid_df['MambaSL (trainlossonly)'] - valid_df[c]), alternative='greater')
    wilcoxon_res.append((c, res.pvalue))

df_wilcoxon = pd.DataFrame(wilcoxon_res, columns=['Classifier', 'p-value']).set_index('Classifier')
df_wilcoxon.T
```
