### Get the min/max sequence length of irregular datasets

```python
import pandas as pd
import os
```

```python
for fname in sorted(os.listdir('.')):
    if fname.endswith('.pkl'):
        df = pd.read_pickle(fname)
        print(fname)
        print(df.groupby(df.index).count().min().iloc[0], 
              df.groupby(df.index).count().max().iloc[0])
```
