```python
import pandas as pd
import plotly.express as px
from plotly.io import write_image
from sklearn.manifold import TSNE
```

```python
# dataset loading
df = pd.read_csv('all-results.csv', index_col=0, header=0).T
```

### t-SNE

```python
for i, p in enumerate([3,5,10]): 
    # t-SNE fitting
    tsne = TSNE(n_components=2, perplexity=p,
                max_iter=int(1e6), random_state=0) 
    X_tsne = tsne.fit_transform(df)

    # t-SNE visualization DataFrame
    df_tsne = pd.DataFrame(X_tsne, columns=['x', 'y'])
    df_tsne['model'] = df.index
    df_tsne['acc'] = df.mean(axis=1).values

    fig = px.scatter(df_tsne, x='x', y='y', text='model', 
                    labels={'x': 'component 1', 'y': 'component 2'},
                    color='acc',
                    color_continuous_scale=px.colors.sequential.Viridis,
                    size='acc',
                    size_max=8,
                    opacity=0.7)
    # fig size
    fig.update_layout(width=800, height=400)
    # title
    fig.update_layout(title=f'Visualization of TSC models using t-SNE (perplexity={p}, metric=euclidean)', 
                        title_x=0.5,  # center the title
                        title_y=0.98,
                        title_font=dict(size=20))
    # tight layout
    fig.update_layout(
        coloraxis_colorbar=dict(
            title="avg.<br>acc",
            title_font=dict(size=13),
            tickmode="array",
            tickvals=list(range(70,80,2)),
            lenmode="fraction",
            thickness=20,
            len=1,
            yanchor="middle",
            y=0.5,
            xanchor="left",
            x=1.0,
        ),
    )
    fig.update_layout(margin=dict(l=0, r=0, t=30, b=15))
    fig.update_xaxes(title_standoff=7)
    fig.update_yaxes(title_standoff=7)

    # text size
    fig.update_traces(textfont_size=18)
    fig.update_traces(textposition='bottom center')
    fig.update_layout(showlegend=False)
    fig.show()

    # # save the figure
    # write_image(fig, f'figures_tmp/tsne_perplexity_{p}.pdf', width=800, height=400, scale=2, format='pdf')
```

```python
import plotly.graph_objects as go

color_map = {
    "Non-DL": "#B0B0B0",            # 회색
    "MLP-based": "#7D68B1",           # 오렌지
    "CNN-based": "#D5B85B",          # 스카이 블루
    "Transformer-based": "#2C5AA0",  # 다크 블루
    "Shaplet+CNN": "#5BB3C5",        # 청록
    "Mamba-based": "#F2A900",        # 노랑
}



# 메달 이모지
medals = ["🥇", "🥈", "🥉"]

# subplot 그리기
from plotly.subplots import make_subplots

fig = make_subplots(rows=1, cols=1)

for i, c in enumerate(df['categories'].unique()):
    df_subset = df[df['categories'] == c].copy()
    fig.add_trace(
            go.Bar(
                x=df_subset.index,
                y=df_subset.iloc[:, :10].mean(axis=1),
                marker_color=[color_map[c] for c in df_subset['categories']],
                marker_pattern_shape=['x'] * len(df_subset),
                text=[f"{v:.1f}" for v in df_subset.iloc[:, :10].mean(axis=1)],
                textposition="outside",
                hovertext=[f"<b>{df_subset.index[i]}</b><br>{c} <br>{df_subset.iloc[i, :10].mean():.2f}" for i in range(len(df_subset))],
                hoverinfo="text",
                name="10 dataset\n(EC-UW)",
                legendgroup=c,
                legendgrouptitle_text=c,
                
            )
    )
    fig.add_trace(
            go.Bar(
                x=df_subset.index,
                y=df_subset.iloc[:, :-1].mean(axis=1),
                marker_color=[color_map[c] for c in df_subset['categories']],
                text=[f"{v:.1f}" for v in df_subset.iloc[:, :-1].mean(axis=1)],
                textposition="outside",
                hovertext=[f"<b>{df_subset.index[i]}</b><br>{c} <br>{df_subset.iloc[i, :-1].mean():.2f}" for i in range(len(df_subset))],
                hoverinfo="text",
                name="all dataset",
                legendgroup=c,
                
            )
    )


fig.update_traces(
    marker=dict(line_color="grey", pattern_fillmode="replace"),
    # marker_color=df_subset['categories'].iloc[::2].map(color_map),
    texttemplate='%{text:.1f}',
    textposition='outside',
    hovertemplate='<b>%{x}</b><br>Category: %{marker.color}<br>Accuracy: %{y:.2f}<extra></extra>',
)

fig.update_layout(
    height=700,
    width=1200,
    title_text="Classification Performance of TSC Models",
    barmode="group",
    margin=dict(l=20, r=20, t=30, b=30),
    title_x=0.5,  # center the title
    title_font=dict(size=16),
    xaxis=dict(
        tickmode='array',
        tickvals=list(range(len(df.index))),
        ticktext=df.index,
        tickangle=-45,
        title_font=dict(size=14)
    ),
    yaxis_title="average accuracy (%)",
    legend=dict(
        title_text="Categories",
        title_font=dict(size=14),
        font=dict(size=12),
        orientation="h",
    ),
)

fig.update_yaxes(showticklabels=False, range=[66, 81])  # y축 범위 설정
fig.show()

# save the figure
write_image(fig, f'figures_tmp/legend.pdf', scale=2, format='pdf', width=1200, height=700)
```

```python
# subplot 그리기
from plotly.subplots import make_subplots

fig = make_subplots(rows=1, cols=1)
df_pivot = []
for i, c in enumerate(df['categories'].unique()):
    for model in df[df['categories'] == c].index:
        # print(model)
        df_pivot.append({
            'categories': c,
            'model': model,
            'acctype': 'all dataset',
            'acc': df.loc[model].iloc[:-1].mean(),
        })
        df_pivot.append({
            'categories': c,
            'model': model,
            'acctype': '10 dataset',
            'acc': df.loc[model].iloc[:10].mean(),
        })
df_pivot = pd.DataFrame(df_pivot)

df_pivot.head()
```

```python
# 색상 매핑
color_map = {
    "Non-DL": "#B0B0B0",            # 회색
    "MLP-based": "#7D68B1",           # 오렌지
    "CNN-based": "#D5B85B",          # 스카이 블루
    "Transformer-based": "#2C5AA0",  # 다크 블루
    "Shaplet+CNN": "#5BB3C5",        # 청록
    "Mamba-based": "#F2A900",        # 노랑
}


# 메달 이모지
medals = ["🥇", "🥈", "🥉"]


fig = px.bar(df_pivot, x='model', y='acc', 
             pattern_shape='acctype',
             pattern_shape_sequence=['x', ''],
             text='acc',
             text_auto='.2f',
             category_orders={'model': df_pivot['model'].unique(), 'acctype': ['10 dataset', 'all dataset', ]},
             custom_data=['categories'],
             )

fig.update_traces(
    marker=dict(line_color="grey", pattern_fillmode="replace"),
    marker_color=df_pivot['categories'].iloc[::2].map(color_map),
    texttemplate='%{text:.1f}',
    textposition='outside',
    textfont=dict(size=10),
    hovertemplate='<b>%{x}</b><br>Category: %{customdata[0]}<br>Accuracy: %{y:.6f}<extra></extra>',
)
# update names
fig.for_each_trace(
    lambda t: t.update(name=t.name.replace('10 dataset', '10 dataset\n(EC-UW)').replace('all dataset', 'all dataset')) 
    if t.name in ['10 dataset', 'all dataset'] else t
)



# add medal emojis to the top of the bars
top3 = df_pivot[(df_pivot['acctype'] == 'all dataset')].nlargest(3, 'acc')
for i, (j, row) in enumerate(top3.iterrows()):
    # find x position of the bar
    x_pos = df_pivot[df_pivot['model'] == row['model']].index[0] // 2
    fig.add_annotation(
        x=x_pos + 0.2,
        y=row['acc'] + 1.6,  # Adjust the y position to place the emoji above the bar
        text=medals[i],
        showarrow=False,
        font=dict(size=20, family="Noto Color Emoji"),  # Adjust the size of the emoji
        align='center',
        bgcolor='rgba(0,0,0,0)',  # Transparent background
    )

top3_10 = df_pivot[(df_pivot['acctype'] == '10 dataset')].nlargest(3, 'acc')
for i, (j, row) in enumerate(top3_10.iterrows()):
    # find x position of the bar
    x_pos = df_pivot[df_pivot['model'] == row['model']].index[0] // 2
    fig.add_annotation(
        x=x_pos - 0.2,
        y=row['acc'] + 1.6,  # Adjust the y position to place the emoji above the bar
        text=medals[i],
        showarrow=False,
        font=dict(size=20, family="Noto Color Emoji"),  # Adjust the size of the emoji
        align='center',
        bgcolor='rgba(0,0,0,0)',  # Transparent background
    )






fig.update_layout(
    # template="plotly_white",
    paper_bgcolor='rgba(0,0,0,0)',  # 투명 배경
    plot_bgcolor="#E8ECF1",  # 연한 회색 배경
    height=400,
    width=1200,
    barmode='group',
    bargap=0.2,  # 바 간격 조절
    bargroupgap=0.0,  # 그룹 간격 조절
    # title_text="Classification Performance of TSC Models",
    margin=dict(l=20, r=20, t=30, b=10),
    title_x=0.5,  # center the title
    title_font=dict(size=16),
    yaxis_title="average accuracy (%)",
    # legend=dict(
    #     title_text="Categories",
    #     title_font=dict(size=14),
    #     font=dict(size=12)
    # ),
    showlegend=False,  # Hide legend
)
fig.update_xaxes(
    title_text="",
    tickangle=-30,
    tickfont=dict(size=15)
)
fig.update_yaxes(showticklabels=False, range=[66, 81.5])  # y축 범위 설정
fig.show()

# save the figure
write_image(fig, f'figures_tmp/barplot (acc).pdf', scale=2, format='pdf', width=1200, height=400)
```
