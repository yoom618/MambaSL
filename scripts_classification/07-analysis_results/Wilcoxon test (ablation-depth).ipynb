```python
import numpy as np
import pandas as pd
pd.set_option('display.float_format', lambda x: '%.13f' % x)
pd.set_option('display.max_columns', None)

from scipy.stats import wilcoxon
```

```python
df = pd.read_excel('all-depth-ablation.xlsx', index_col=0)
df.head()
```

```python
df_norm = ((df.T - df.mean(axis=1)) / df.std(axis=1)).T
df_norm.fillna(0, inplace=True)
wilcoxon_res = list()
for c in df.columns:
    if c == '1 (proposed)':
        continue
    res = wilcoxon((df_norm['1 (proposed)'] - df_norm[c]), alternative='greater')
    wilcoxon_res.append((c, res.pvalue))

df_wilcoxon = pd.DataFrame(wilcoxon_res, columns=['Classifier', 'p-value']).set_index('Classifier')
df_wilcoxon.T
```
